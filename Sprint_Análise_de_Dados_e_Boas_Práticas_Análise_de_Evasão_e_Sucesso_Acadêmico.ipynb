## Introdução

Este trabalho tem como objetivo analisar um dataset que investiga fatores associados à evasão e ao sucesso acadêmico de estudantes do ensino superior. A base contém informações demográficas, acadêmicas, socioeconômicas e institucionais dos alunos, bem como dados relacionados ao desempenho acadêmico ao longo dos primeiros semestres.

A escolha desse dataset se justifica pela relevância do tema, uma vez que a evasão acadêmica é um problema importante nas instituições de ensino, com impactos sociais e econômicos. Além disso, o dataset permite explorar relações entre características dos estudantes e sua permanência ou desempenho acadêmico.


## Definição do problema

O problema analisado neste trabalho é compreender de que forma características dos estudantes, como perfil demográfico, contexto socioeconômico e desempenho acadêmico inicial estão associadas à evasão e ao sucesso acadêmico.

Busca-se investigar se variáveis como notas de ingresso, desempenho nos primeiros semestres, situação financeira, idade e características familiares apresentam relação com a situação final do estudante.

Trata-se de um problema de aprendizado supervisionado, uma vez que o dataset apresenta uma variável alvo (`Target`) que indica o desfecho do estudante (evasão, permanência ou sucesso).

Como hipótese inicial, considera-se que estudantes com melhor desempenho acadêmico inicial e maior estabilidade financeira apresentam maior probabilidade de sucesso. Por outro lado, dificuldades acadêmicas e fatores socioeconômicos desfavoráveis podem estar associados à evasão.

# Pré-Processamento

Objetivo: preparar a base para análises futuras e possíveis aplicações de modelagem.

## Etapa 1. Separação entre variáveis categóricas e numéricas

```python
# Separação entre variáveis categóricas e numéricas
```

```python

variaveis_categoricas = [
    'Marital status',
    'Application mode',
    'Application order',
    'Course',
    'Daytime/evening attendance',
    'Previous qualification',
    'Nacionality',
    "Mother's qualification",
    "Father's qualification",
    "Mother's occupation",
    "Father's occupation",
    'Displaced',
    'Educational special needs',
    'Debtor',
    'Tuition fees up to date',
    'Gender',
    'Scholarship holder',
    'International'
]

variaveis_numericas = [
    'Previous qualification (grade)',
    'Admission grade',
    'Age at enrollment',
    'Curricular units 1st sem (credited)',
    'Curricular units 1st sem (enrolled)',
    'Curricular units 1st sem (evaluations)',
    'Curricular units 1st sem (approved)',
    'Curricular units 1st sem (grade)',
    'Curricular units 1st sem (without evaluations)',
    'Curricular units 2nd sem (credited)',
    'Curricular units 2nd sem (enrolled)',
    'Curricular units 2nd sem (evaluations)',
    'Curricular units 2nd sem (approved)',
    'Curricular units 2nd sem (grade)',
    'Curricular units 2nd sem (without evaluations)',
    'Unemployment rate',
    'Inflation rate',
    'GDP'
]
```

Embora diversas variáveis estejam representadas numericamente, parte delas corresponde a categorias codificadas. Dessa forma, foi realizada a separação entre variáveis categóricas e numéricas, a fim de permitir tratamento adequado na etapa de pré-processamento.

## Etapa 2. Conversão de tipo

```python
# Conversão de variáveis categóricas
```

```python

df_pre = df.copy()

for col in variaveis_categoricas:
    df_pre[col] = df_pre[col].astype('category')

df_pre.info()
```

Após a conversão dos tipos, observa-se que as variáveis categóricas codificadas numericamente passaram a ser tratadas como `category`, enquanto as variáveis quantitativas permaneceram como `int64` ou `float64`.

Essa etapa é importante porque evita interpretações incorretas de códigos numéricos como se fossem medidas quantitativas. Dessa forma, atributos como curso, nacionalidade, gênero e modalidade de ingresso passam a ser tratados de forma mais adequada nas etapas seguintes do pré-processamento.

Além disso, a redução do uso de memória após a conversão indica maior eficiência no armazenamento da base, o que também é um efeito positivo dessa transformação.

```python
# Ajeitando a categoria do Target
```

```python
df_pre['Target'] = df_pre['Target'].astype('category')
df_pre.info()
```

Por fim, a variável alvo `Target` também foi convertida para o tipo categórico, uma vez que representa classes nominais relacionadas à situação final do estudante.

## Etapa 3. One-hot-encoding

```python
# One-hot encoding das variáveis categóricas
```

```python
df_encoded = pd.get_dummies(df_pre, columns=variaveis_categoricas, drop_first=True)
df_encoded.head()
```

Após a aplicação do one-hot encoding, observa-se um aumento significativo no número de colunas do dataset, passando de 37 para 245 variáveis.

Esse aumento ocorre porque cada categoria das variáveis categóricas é transformada em uma coluna binária, representando a presença ou ausência daquela categoria em cada observação.

Variáveis com grande número de categorias, como curso, ocupação dos pais e nacionalidade, contribuem de forma mais expressiva para esse crescimento dimensional.

Embora essa transformação seja importante para permitir o uso de algoritmos de aprendizado de máquina, ela também pode aumentar a complexidade do modelo e gerar redundância de informação, sendo necessário considerar estratégias adicionais em etapas futuras, como seleção de atributos.

De modo geral, a transformação foi realizada com sucesso, resultando em uma base totalmente numérica e adequada para análises quantitativas e modelagem.

Obs: A opção `drop_first=True` foi utilizada para evitar redundância entre categorias, reduzindo o risco de multicolinearidade entre variáveis derivadas de um mesmo atributo.

## Etapa 4. Padronização

```python
# Padronização das variáveis numéricas
```

```python
from sklearn.preprocessing import StandardScaler

df_padronizado = df_encoded.copy()

scaler = StandardScaler()
df_padronizado[variaveis_numericas] = scaler.fit_transform(df_padronizado[variaveis_numericas])

df_padronizado[variaveis_numericas].head()
```

Com o objetivo de reduzir o impacto das diferentes escalas entre as variáveis numéricas, foi aplicada a técnica de padronização aos atributos quantitativos do dataset.

A padronização transforma os dados de forma que cada variável passe a apresentar média igual a zero e desvio-padrão igual a um. Como resultado, os valores transformados podem assumir números negativos e positivos, dependendo de sua posição em relação à média da variável original.

Essa etapa é especialmente importante neste contexto, pois o dataset contém variáveis com escalas bastante distintas, como notas (em diferentes intervalos), contagens de unidades curriculares e indicadores macroeconômicos. Sem esse tratamento, variáveis com maior magnitude poderiam exercer influência desproporcional em análises e modelos.

Dessa forma, a padronização contribui para tornar as variáveis comparáveis entre si e prepara a base para possíveis aplicações futuras de técnicas de aprendizado de máquina.

## Etapa 5. Normalização

```python
# Normalização das variáveis numéricas
```

```python
from sklearn.preprocessing import MinMaxScaler

df_normalizado = df_encoded.copy()

normalizador = MinMaxScaler()
df_normalizado[variaveis_numericas] = normalizador.fit_transform(df_normalizado[variaveis_numericas])

df_normalizado[variaveis_numericas].head()
```

Com o objetivo de reduzir diferenças de escala entre as variáveis numéricas, foi aplicada a técnica de normalização aos atributos quantitativos do dataset.

A normalização reescala os valores das variáveis para um intervalo fixo, geralmente entre 0 e 1. Como resultado, todos os atributos passam a estar na mesma escala, mantendo a proporcionalidade entre os valores originais.

Essa transformação é especialmente relevante neste dataset, pois há variáveis com magnitudes bastante distintas, como notas, contagens de unidades curriculares e indicadores macroeconômicos. Sem esse ajuste, variáveis com valores maiores poderiam exercer influência desproporcional em análises e modelos.

Observa-se que, após a normalização, todos os valores passaram a estar contidos no intervalo [0, 1], facilitando a comparabilidade entre os atributos.

Dessa forma, a normalização contribui para uma representação mais equilibrada dos dados, sendo particularmente útil em métodos que dependem da magnitude relativa das variáveis.

Diferentemente da padronização, a normalização não centraliza os dados em torno da média, mas ajusta diretamente sua escala, preservando a forma da distribuição original.

## Etapa 6. Discretização opcional da idade

```python
# Discretização da idade
```

```python

df_discretizado = df.copy()

df_discretizado['Age group'] = pd.cut(
    df_discretizado['Age at enrollment'],
    bins=[15, 20, 25, 30, 40, 80],
    labels=['até 20', '21-25', '26-30', '31-40', '40+']
)

df_discretizado[['Age at enrollment', 'Age group']].head()
```


Como forma de facilitar a interpretação e permitir análises comparativas entre grupos, a variável `Age at enrollment` foi discretizada em faixas etárias.

Foram definidos intervalos representando diferentes momentos da vida acadêmica, agrupando os estudantes em categorias como até 20 anos, 21 a 25 anos, 26 a 30 anos, 31 a 40 anos e acima de 40 anos.

Essa transformação permite reduzir a granularidade da variável original, tornando mais simples identificar padrões relacionados à idade, especialmente em relação à evasão e ao desempenho acadêmico.

Cabe destacar que a discretização não substitui a variável contínua, mas cria uma nova representação complementar que pode ser útil em análises específicas.

 A discretização pode ser particularmente útil quando se busca capturar efeitos não lineares da idade sobre o comportamento dos estudantes, que nem sempre são evidentes ao analisar a variável em sua forma contínua.
